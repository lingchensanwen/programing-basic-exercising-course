```python
import math
f = [lambda x,y: x - x**3 -y, lambda x,y:x]
start = [2,1]
def ndsolve(f,start,tstart,tend):
    nt = tstart#record time
    cx = start[0]#current x
    cy = start[1]#current y
    nlist = []
    while(nt <= tend):
        nx = cx + 0.01*f[0](cx,cy)#new x
        cx = nx
        ny = cy + 0.01*f[1](cx,cy)
        cy = ny
        nt = nt + 0.01#new time
        #print([nt,cx,cy])
        nlist.append([nt,cx,cy])
    return nlist
```

```python
fs = [lambda x,y: x - x**3 - y, lambda x,y: x ]
traj = ndsolve(fs, [2,1], 0, 30)
print(traj)
```

```python
import matplotlib.pyplot as plt
ttr, xtr, ytr = zip(*traj)
plt.plot(xtr, ytr)
plt.show()
```

```python
def logistic(a):
    x = 0.1
    while True:
        xnext = a * x * (1 - x)
        yield xnext
        x = xnext
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
am = [4.0 * n / 100 for n in range(100)]
for a in am:
    g = logistic(a)
    traj = [next(g) for i in range(100)]
    i = 0
    x = []
    while i <len(traj):
        x.append(a)
        i = i + 1
    plt.scatter(x,traj,s=0.1)
```

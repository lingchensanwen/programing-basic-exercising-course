```python
import numpy as np
import cmath
import math
import matplotlib.pyplot as plt
%matplotlib inline
arr = np.random.randn(2000,2000)
```

```python
m = np.asmatrix(arr)/math.sqrt(2000)
```

```python
a,b = np.linalg.eig(m)
```

```python
x = a.real
y = a.imag
```

```python
plt.scatter(x,y)
```
